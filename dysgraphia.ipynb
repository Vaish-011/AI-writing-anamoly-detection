```python
import kagglehub
import shutil
import os

# Download dataset
path = kagglehub.dataset_download("amrmohamedahemd/dysgraphia1")

# Copy all files from KaggleHub cache to current directory
dest = os.getcwd()
for item in os.listdir(path):
    s = os.path.join(path, item)
    d = os.path.join(dest, item)
    if os.path.isdir(s):
        shutil.copytree(s, d, dirs_exist_ok=True)
    else:
        shutil.copy2(s, d)

print("Dataset copied to:", dest)
```
